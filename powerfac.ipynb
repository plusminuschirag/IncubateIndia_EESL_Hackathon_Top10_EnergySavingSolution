```python
import pandas as pd
import matplotlib.pyplot as plt
import re
import seaborn as sns
```

```python
df = pd.read_excel('datafiltered.xls',sheet_name='Sheet13')
```

```python
df.head()
```

```python
date_time_cleaned = []
for d in df['Date &Time']:
    result = re.match(r'\d\d-.*-\d\d\d\d (\d\d:\d\d)(:\d\d)',d)
    date_time_cleaned.append(result.group(1))
```

```python
df['Date &Time'] = date_time_cleaned
```
