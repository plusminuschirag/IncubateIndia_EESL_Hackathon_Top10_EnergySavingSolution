```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('finalfile.xls',sheet_name='Sheet1',header=None)
df.head()
```

```python
df.drop([1,2,3,4],inplace=True,axis=1)
```

```python
df.head()
df = df.rename(columns={0:'Captured Date and Time',5:'Active Energy Import KWH',6:'Active Energy Export KWH',7:'Apparent_Energy_Import_KVAh',8:'Apparent Energy Export KVAh',9:'Average Voltage',10:'Average Current',11:'Neutral Current'})
```

```python
df.head()
print(max(df['Active Energy Import KWH']))
```

```python
df['5'] = 5
df['Cost'] = 0
df.loc[df['Active Energy Import KWH']> 0.8, 'Cost'] = 5
df.loc[df['Active Energy Import KWH'].between(0.6,0.8),'Cost'] = 6
df.loc[df['Active Energy Import KWH'].between(0.4,0.6),'Cost'] = 7
df.loc[df['Active Energy Import KWH']< 0.4,'Cost'] = 8
df.loc[df['Cost']>0]
```

```python
plt.figure(figsize=(20,10))
plt.plot(df['Captured Date and Time'],df['Active Energy Import KWH']/df['Apparent Energy Import KVAh'])
```
